```python
import os
from dotenv import load_dotenv

# Load API key from .env file (if it exists)
load_dotenv()

# Configuration
DEFAULT_MODEL = os.getenv("DEFAULT_MODEL")
DEFAULT_TEMPERATURE = 0.7
DEFAULT_MAX_TOKENS = 500

print(f"‚úÖ Using model: {DEFAULT_MODEL}")
```

## üõ†Ô∏è Helper Functions

Let's create functions to work with reasoning models.

```python
from litellm import completion

import litellm
import logging

litellm.suppress_debug_info = True
logging.getLogger("litellm").setLevel(logging.CRITICAL)

def solve_with_reasoning(problem: str, temperature: float = DEFAULT_TEMPERATURE, max_tokens: float = DEFAULT_MAX_TOKENS) -> str:
    """
    Solve a problem using chain-of-thought reasoning.
    
    Args:
        problem: The problem to solve
        temperature: Lower = more focused, higher = more creative
    
    Returns:
        Solution with reasoning steps
    """
    # Add chain-of-thought prompt
    prompt = f"""{problem}

Please solve this step-by-step:
1. First, understand what is being asked
2. Break down the problem into steps
3. Solve each step
4. Verify your answer
5. Provide the final answer

Show your reasoning for each step."""
    
    response = completion(
        model=DEFAULT_MODEL,
        messages=[{"role": "user", "content": prompt}],
        temperature=temperature,
        max_tokens=max_tokens
    )
    
    return response.choices[0].message.content

print("‚úÖ Helper functions loaded!")
```

## ü§î Part 1: Understanding Reasoning Models

**What makes reasoning models special?**

Regular AI models predict the next word based on patterns. Reasoning models:
- **Think before answering**: They generate internal reasoning steps
- **Show their work**: You can see how they arrived at the answer
- **Self-verify**: They check their own answers
- **Handle complexity**: Better at multi-step problems

**Real-World Analogy:** It's like the difference between:
- **Regular AI**: A student who memorized answers
- **Reasoning AI**: A student who understands the concepts and can explain their thinking

### üí° Key Insight

**Chain-of-Thought (CoT) Prompting** is the key technique. Instead of asking "What's the answer?", you ask "Solve this step-by-step and show your work."

This simple change dramatically improves accuracy on complex problems!

## üßÆ Part 2: Simple Reasoning Test

**Scenario:** A student asks your tutoring app: "If it takes 5 minutes to boil one egg, how long does it take to boil three eggs?"

This is a classic trick question! Let's see how reasoning helps.

```python
problem = "If it takes 5 minutes to boil one egg, how long does it take to boil three eggs?"

print(f"[Problem:] {problem}\n")

solution = solve_with_reasoning(problem)

print(f"‚úì Solution with Reasoning:\n")
print(solution)
```

### ‚ùì Discussion Question #1

Did the AI fall for the trick? The correct answer is **5 minutes** (you can boil all three eggs at the same time!).

Notice how the reasoning process helps the AI think through the problem logically instead of just pattern-matching to "3 eggs √ó 5 minutes = 15 minutes".

## üî¢ Part 3: Complex Math Problem

**Scenario:** A student needs help with a challenging math problem. Regular AI might get it wrong, but reasoning models can work through it step-by-step.

**What this means:** Instead of guessing, the AI will:
1. Identify what needs to be calculated
2. Break it into smaller steps
3. Solve each step
4. Verify the final answer

```python
problem = "What is the sum of the first 10 prime numbers? Keep the answer concise under 100 words."

print(f"[Problem:] {problem}\n")

solution = solve_with_reasoning(problem, temperature=0.3)  # Lower temp for math

print(f"‚úì Solution:")
print(solution)
```

### üí° Key Insight

For math problems, use **lower temperature** (0.2-0.3) to get more focused, deterministic reasoning. Higher temperature can lead to creative but incorrect solutions!

## üêõ Part 4: Code Debugging with Reasoning

**Scenario:** A student wrote code that has a bug. Instead of just fixing it, your tutoring app should explain what's wrong and why.

**Why reasoning helps:** The AI needs to:
1. Understand what the code is supposed to do
2. Trace through the logic
3. Identify where it goes wrong
4. Explain the fix

```python
buggy_code = '''
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-3)  # Bug here!
'''

debug_prompt = f"""Find and explain the bug in this code:
{buggy_code}

1. First, explain what the code is trying to do
2. Trace through an example (e.g., fibonacci(5))
3. Identify where the bug is
4. Explain why it's wrong
5. Provide the corrected version

Show your reasoning for each step. Keep it concise under 100 words."""

print(f"[Buggy Code:]")
print(buggy_code)
print()

solution = solve_with_reasoning(debug_prompt, temperature=0.3, max_tokens=1000)

print(f"‚úì Debug Analysis:\n")
print(solution)
```

## üß© Part 5: Logical Reasoning Puzzle

**Scenario:** Your app needs to solve logic puzzles that require careful step-by-step thinking.

**What this means:** Problems like "If all A are B, and all B are C, what can we conclude about A and C?" require logical deduction, not just pattern matching.

```python
logic_puzzle = """
Three friends - Alice, Bob, and Carol - are standing in a line.
- Alice is not first
- Bob is not last
- Carol is not in the middle

What is the order of the three friends? Make sure answer is concise under 200 words.
"""

print(f"[Logic Puzzle:]")
print(logic_puzzle)

solution = solve_with_reasoning(logic_puzzle, temperature=0.3)

print(f"‚úì Solution:\n")
print(solution)
```

### ‚ùì Discussion Question #2

Notice how the AI needs to:
1. List the constraints
2. Try different arrangements
3. Check each against the constraints
4. Find the valid solution

This is exactly how a human would solve it! What other types of logic puzzles could benefit from this approach?

## üíº Real-World Applications

### 1. Math Tutoring Bot
**What it does:** Helps students learn by showing step-by-step solutions

**How to build it:**
- Use chain-of-thought prompting for all problems
- Show reasoning steps, not just answers
- Ask follow-up questions to check understanding

```python
# Example: Math Tutoring Bot
def math_tutor(problem: str) -> str:
    """Tutoring bot that explains step-by-step."""
    tutor_prompt = f"""
You are a patient math tutor. A student asks:

{problem}

Provide a step-by-step explanation that helps them learn, not just the answer.
Use simple language and encourage them to think through each step. Make the answer under 200 words.
"""
    
    response = completion(
        model=DEFAULT_MODEL,
        messages=[{"role": "user", "content": tutor_prompt}],
        temperature=0.5,
        max_tokens=DEFAULT_MAX_TOKENS
    )
    
    return response.choices[0].message.content

# Test it
student_question = "How do I solve: 2x + 5 = 15?"
print(f"[Student:] {student_question}\n")

tutor_response = math_tutor(student_question)
print(f"‚úì Tutor:\n")
print(tutor_response)
```

### 2. Code Review Assistant
**What it does:** Reviews code and explains potential issues with reasoning

**What this means:** Instead of just saying "This is wrong," it explains:
- What the code does
- What could go wrong
- Why it's a problem
- How to fix it

```python
# Example: Code Review Assistant
code_to_review = '''
def divide_list(numbers, divisor):
    result = []
    for num in numbers:
        result.append(num / divisor)
    return result
'''

review_prompt = f"""
Review this code for potential issues:
{code_to_review}

Think through:
1. What does this code do?
2. What inputs could cause problems?
3. What are the edge cases?
4. How should it be improved?

Provide a thorough code review with reasoning.
"""

print(f"[Code to Review:]")
print(code_to_review)
print()

review = solve_with_reasoning(review_prompt, temperature=0.3, max_tokens=2000)
print(f"‚úì Code Review:")
print(review)
```

### 3. Decision Analysis Tool
**What it does:** Helps users make complex decisions by analyzing pros/cons

**How to build it:**
- Present the decision scenario
- Ask AI to list all factors
- Weigh pros and cons systematically
- Provide a reasoned recommendation

```python
# Example: Decision Analysis
decision = """
I'm deciding between two job offers:

Job A:
- Salary: $120,000
- Location: San Francisco (high cost of living)
- Startup with 20 people
- Equity: 0.5%
- Exciting product but risky

Job B:
- Salary: $100,000
- Location: Austin (moderate cost of living)
- Established company with 500 people
- No equity
- Stable but less exciting

Help me think through this decision.
"""

analysis_prompt = f"""
{decision}

Analyze this decision step-by-step:
1. List all important factors
2. Compare each factor
3. Consider long-term implications
4. Identify what matters most
5. Provide a reasoned recommendation

Keep the analysis per step concise.
"""

print(f"[Decision Scenario:]")
print(decision)

analysis = solve_with_reasoning(analysis_prompt, temperature=0.5, max_tokens=2000)
print(f"‚úì Analysis:\n")
print(analysis)
```

## üéì Key Takeaways

Congratulations! You now understand reasoning models:

1. **Chain-of-Thought** - Asking AI to show its work improves accuracy
2. **Step-by-Step** - Breaking problems into steps helps solve complex tasks
3. **Self-Verification** - AI can check its own answers
4. **Transparency** - Seeing reasoning helps trust and debug
5. **Real-world Power** - Tutoring, debugging, decision analysis

### üîë Best Practices:

| Problem Type | Temperature | Max Tokens | Reasoning Depth |
|--------------|-------------|------------|----------------|
| Math | Low (0.2-0.3) | 2000+ | High - show all steps |
| Logic Puzzles | Low (0.3) | 1500+ | High - systematic |
| Code Debugging | Low (0.3) | 2000+ | High - trace execution |
| Decision Analysis | Medium (0.5) | 2000+ | Medium - balanced |
| Creative Problem | Higher (0.7) | 1500+ | Medium - flexible |

### ‚ö†Ô∏è Important Considerations:

- **Cost**: Reasoning models use more tokens (longer responses)
- **Speed**: Step-by-step reasoning takes longer
- **Not always needed**: Simple questions don't need complex reasoning
- **Verification**: AI can still make mistakes - always verify critical answers
